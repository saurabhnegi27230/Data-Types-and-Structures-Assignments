**Data Types and Structures Questions And Answer**

1. What are data structures, and why are they important?

Answer:
Data structures are ways to organize and store data so we can use it efficiently. They help us manage data, search it faster, and perform operations like adding, deleting, or updating data easily.

2. Explain the difference between mutable and immutable data types with examples
Answer:
Mutable means we can change the data after creating it. Example: list → [1, 2, 3]

Immutable means we cannot change the data once created. Example: tuple → (1, 2, 3)

3. What are the main differences between lists and tuples in Python?

Answer:
Lists are mutable (can change), tuples are immutable (cannot change).

Lists use [ ], tuples use ( ).

Tuples are faster and safer when you don’t want data to change.

4. Describe how dictionaries store data

Answer:
Dictionaries store data as key-value pairs. Each key points to a value, like:

{"name": "Alice", "age": 25}
5. Why would you use a set instead of a list in Python?

Answer:
Use a set when you want to store only unique items and don’t care about order. Sets automatically remove duplicates.

6. What is a string in Python, and how is it different from a list?

Answer:
A string is a group of characters (like "hello").

Strings are immutable, lists are mutable.

You can change list items, but not characters in a string.

7. How do tuples ensure data integrity in Python?

Answer:
Since tuples cannot be changed, they help protect data from being accidentally modified. That’s why they are used when the data should stay constant.

8. What is a hash table, and how does it relate to dictionaries in Python?

Answer:
A hash table is a data structure that stores data using a key and hash function.
Dictionaries in Python use hash tables internally to store key-value pairs, allowing fast lookup.

9. Can lists contain different data types in Python?

Answer:
Yes, lists can hold different types like numbers, strings, or even other lists.
Example:
[1, "hello", [2, 3]]
10. Explain why strings are immutable in Python

Answer:
Strings are immutable to make them safe and efficient. This helps avoid accidental changes and improves performance during repeated use.

11. What advantages do dictionaries offer over lists for certain tasks?

Answer:
Dictionaries allow faster lookups using keys, while lists need to search item by item.
For example, finding a student by name is faster in a dictionary.

12. Describe a scenario where using a tuple would be preferable over a list

Answer:
When you want to store constant values that shouldn’t change, like days of the week:
("Monday", "Tuesday", "Wednesday", ...)
13. How do sets handle duplicate values in Python?

Answer:
Sets automatically remove duplicates.
Example:
set([1, 2, 2, 3]) → {1, 2, 3}
14. How does the “in” keyword work differently for lists and dictionaries?

Answer:

In a list, in checks values.

In a dictionary, in checks only keys, not values.

15. Can you modify the elements of a tuple? Explain why or why not
Answer:
No, you cannot modify tuple elements because tuples are immutable. Once created, the data stays fixed.

16. What is a nested dictionary, and give an example of its use case

Answer:
A nested dictionary is a dictionary inside another dictionary.
Example:

student = {
  "name": "Amit",
  "marks": {"math": 90, "science": 85}
}
Use case: storing data for multiple subjects of a student.

17. Describe the time complexity of accessing elements in a dictionary

Answer:
Accessing items in a dictionary is O(1) on average, which means it’s very fast, regardless of how big the dictionary is.

18. In what situations are lists preferred over dictionaries?

Answer:
Use lists when:

Order matters.

You need to store simple collections of items (e.g., names, numbers).

You don’t need key-value pairing.

19. Why are dictionaries considered unordered, and how does that affect data retrieval?

Answer:
 dictionaries did not keep order. Now they do, but still, we usually use them for key-based access, not for order.

20. Explain the difference between a list and a dictionary in terms of data retrieval.

Answer:

In a list, you get data by index (number position).

In a dictionary, you get data by key (name/label), which is faster and more readable.
